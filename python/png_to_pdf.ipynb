```python
import os
import img2pdf

def convert_png_to_pdf(input_folder, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of PNG files in the input folder
    png_files = [f for f in os.listdir(input_folder) if f.endswith('.png')]

    for png_file in png_files:
        # Construct the paths
        input_path = os.path.join(input_folder, png_file)
        output_path = os.path.join(output_folder, os.path.splitext(png_file)[0] + '.pdf')

        # Convert PNG to PDF using img2pdf
        with open(output_path, 'wb') as pdf_file:
            pdf_file.write(img2pdf.convert(input_path))

if __name__ == "__main__":
    input_folder = "/Users/samrat/Desktop/shear_arcade_draft2/"
    output_folder = "/Users/samrat/Desktop/shear_arcade_draft2/pdf2/"

    convert_png_to_pdf(input_folder, output_folder)
```
